 - sns.distplot(df[x], bins=20): Uses Seaborn's distplot function to plot the distribution of the feature x from the DataFrame df. The bins=20 argument  - specifies that the histogram should be divided into 20 bins.
 - plt.title("Displot of {}".format(x)): Sets the title of the current subplot. The title is formatted to include the feature name x.

#### Consider the relationship between the age and spending score and cluster data accordingly

```python
X1 = df.loc[:, ["Age", "Spending Score (1-100)"]].values

from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    k_means = KMeans(n_clusters=i, init="k-means++")
    k_means.fit(X1)
    wcss.append(k_means.inertia_)

plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

Inside the loop:
1. We create an instance of KMeans with n_clusters=i (the number of clusters) and init="k-means++" (a method to start with better cluster centers so the algorithm finds good clusters faster).
2. We fit the KMeans model to the data X1.
3. We calculate the inertia (WCSS) of the fitted model and append it to the wcss list.

Here's a simple explanation of WCSS:

 - Within-Cluster: WCSS focuses on the distance between data points within the same cluster.

 - Sum of Squares: For each cluster, we calculate the squared distance of each data point to the centroid (center) of that cluster. Then, we add up these squared distances for all data points within the cluster.

So, WCSS is the sum of the squared distances of each data point in all clusters to their respective centroids.

In practical terms, a lower WCSS value indicates that the data points within each cluster are closer to their centroid, meaning the clusters are more compact and well-defined. This is desirable because it suggests that the clustering algorithm has effectively grouped similar data points together.

 > When you fit a KMeans model to your data, the algorithm assigns data points to clusters and computes the centroids of these clusters. The .inertia_ attribute then calculates the sum of the squared distances from each data point to the centroid of the cluster it belongs to.

```python
# let's consider no. of clusters that we want to create and the optimum for our data is equal to 4
k_means = KMeans(n_clusters = 4)
labels = k_means.fit_predict(X1)
print(labels)
```

```python
# centroids
print(k_means.cluster_centers_)
```

```python
# visualizing clusters on graph
scatter = plt.scatter(X1[:,0], X1[:,1], c=k_means.labels_, cmap = "rainbow")
# plotting centroids
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], color = "black")
plt.title("Clusters of customers")
plt.xlabel("Age")
plt.ylabel("Spending Score(1-100)")
#color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Cluster')
```

```python
X2 = df.loc[:, ["Annual Income (k$)", "Spending Score (1-100)"]].values

from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    k_means = KMeans(n_clusters=i, init="k-means++")
    k_means.fit(X2)
    wcss.append(k_means.inertia_)

plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
clusters = k_means.fit_predict(X3)
df["label"] = clusters 

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (20,10))
ax = fig.add_subplot(111, projection = "3d")
ax.scatter(df.Age[df.label == 0], df["Annual Income (k$)"][df.label == 0], df["Spending Score (1-100)"][df.label == 0], c="skyblue", s=60)
ax.scatter(df.Age[df.label == 1], df["Annual Income (k$)"][df.label == 1], df["Spending Score (1-100)"][df.label == 1], c="salmon", s=60)
ax.scatter(df.Age[df.label == 2], df["Annual Income (k$)"][df.label == 2], df["Spending Score (1-100)"][df.label == 2], c="lightgreen", s=60)
ax.scatter(df.Age[df.label == 3], df["Annual Income (k$)"][df.label == 3], df["Spending Score (1-100)"][df.label == 3], c="orange", s=60)
ax.scatter(df.Age[df.label == 4], df["Annual Income (k$)"][df.label == 4], df["Spending Score (1-100)"][df.label == 4], c="purple", s=60)
ax.view_init(30, 185)

# customizing
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel("Spending Score (1-100)")
```

ax.scatter(..., c=color, s=size): This method is used to create scatter plots in 3D.

For each cluster label (0, 1, 2, 3, 4):
 - df.Age[df.label == i], df["Annual Income (k$)"][df.label == i], df["Spending Score (1-100)"][df.label == i]: Filters the DataFrame df to select data points belonging to cluster i based on their respective feature values.
 - c="color": Sets the color of the scatter points for each cluster. Colors ("skyblue", "salmon", "lightgreen", "orange", "purple") are chosen for visual distinction.
 - s=60: Sets the size of the scatter points.

> ax.view_init(elev, azim): Sets the elevation (elev) and azimuth (azim) angles of the plot view. Here, elev=30 sets the elevation angle (tilting the plot 30 degrees up from the horizontal plane), and azim=185 sets the azimuth angle (rotating the plot 185 degrees around the z-axis).
